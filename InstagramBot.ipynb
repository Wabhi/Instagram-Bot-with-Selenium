```python
#this is first................

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import urllib.request
from urllib.request import Request,urlopen
import requests
import os
import time
chrome_options=Options()
chrome_options.add_argument('-incognito')
driver=webdriver.Chrome('C:\\Users\\User\\Desktop\\SeleniumDrivers\\chromedriver.exe',options=chrome_options)

#this is for take url as a input..........
url=input("Enter url: ")
folderName = url.split('/')[-2]
driver.get(url)

#this for maximize the page........
driver.maximize_window()

#this will give title of the page number of followers,posts,following........
m=driver.find_element_by_xpath("//meta[@property='og:description']").get_attribute('content')
s=m.split(',')
print(s[0].strip())
print(s[1].strip())
print(s[2].strip())

#this is for scrolling the page till end.........
p=driver.execute_script("window.scrollTo(0, document.body.scrollHeight);var p=document.body.scrollHeight;return p;")
posts=[]
match=False
while(match==False):
    page_count=p
    time.sleep(4)
    p=driver.execute_script("window.scrollTo(0, document.body.scrollHeight);var p=document.body.scrollHeight;return p;")
    if page_count==p:
        match=True
    print("scroll....")
    #this is for to capture the post's link at a time of each scroll........ 
    links=driver.find_elements_by_tag_name('a')
    for i in links:
        try:
            post=i.get_attribute('href')
        except:pass
        if '/p/' in post:
            #print(post)
            if post not in posts:
                posts.append(post)
        
#this is for capture the link of images and vedios from each post.....
vdo_links=[]
img_links=[]
for post in posts:
    driver.get(post)
    s=driver.current_url.split('/')[-2]
    type_=driver.find_element_by_xpath("//meta[@property='og:type']").get_attribute('content')
    if type_ == 'video':
        vdo_link=driver.find_element_by_xpath("//meta[@property='og:video']").get_attribute('content')
        vdo_links.append(vdo_link)
    else:
        img_link=driver.find_element_by_xpath("//meta[@property='og:image']").get_attribute('content')
        img_links.append(img_link)
#this is for download videos......
for link in vdo_links:
    folder="C:\\Users\\User\\Desktop\\instagram\\"+folderName+"\\video"
    os.makedirs(folder, exist_ok = True)
    data = requests.get(link)    
    file = open(os.path.join(folder, link.split("=")[-1])+'.mp4', "wb")
    file.write(data.content)
#this is for download image......
for link in img_links:
    folder="C:\\Users\\User\\Desktop\\instagram\\"+folderName+"\\image"
    os.makedirs(folder, exist_ok = True)
    data = requests.get(link)    
    file = open(os.path.join(folder, link.split("=")[-1])+'.jpg', "wb")
    file.write(data.content)
```
